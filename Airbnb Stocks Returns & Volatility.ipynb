```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Airbnb data from Yahoo Finance for 10 years
df = yf.download("ABNB", period ='5Y')

# Calculate daily returns and annualized volatility
df['Returns'] = df['Close'].pct_change()
df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

# Plot the S&P 500 prices
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1)
plt.plot(df.index, df['Close'], label='Airbnb Prices', color='blue')
plt.title('AirBnb Prices')
plt.ylabel('Price')
plt.legend()

# Plot the annualized volatility
plt.subplot(2, 1, 2)
plt.plot(df.index, df['Volatility'], label='Annualized Volatility', color='red')
plt.title('AirBnb Annualized Volatility')
plt.ylabel('Annualized Volatility')
plt.xlabel('Year')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 1. Downloading stock data (Apple Inc. as an example)
df = yf.download("ABNB", period = '5Y')

# 2. Calculate daily returns
df['Returns'] = df['Close'].pct_change()

# 3. Calculate rolling annualized volatility (252 trading days = 1 year)
df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

# Drop rows with NaNs created by pct_change and rolling window
df.dropna(inplace=True)

# 4. Plot all three in subplots
plt.figure(figsize=(14, 10))

# Plot stock prices
plt.subplot(3, 1, 1)
plt.plot(df.index, df['Close'], color='blue', label='Stock Price')
plt.title('Stock Price - ABNB')
plt.ylabel('Price')
plt.legend()

# Plot daily returns
plt.subplot(3, 1, 2)
plt.plot(df.index, df['Returns'], color='green', label='Daily Returns')
plt.title('Daily Returns - ABNB')
plt.ylabel('Return')
plt.legend()

# Plot annualized volatility
plt.subplot(3, 1, 3)
plt.plot(df.index, df['Volatility'], color='red', label='Annualized Volatility')
plt.title('Annualized Volatility - ABNB')
plt.ylabel('Volatility')
plt.xlabel('Date')
plt.legend()

plt.tight_layout()
plt.show()
```
